# Previsão de Sobrevivência a Insuficiências Cardíacas

Doenças e sintomas que causam morte podem, em alguns casos, não ter uma cura rápida e imediata, necessitando de diagnóstico o mais rápido possível para um início de tratamento antes que a situação se torne irreversível ou muito arriscada.

Dentre esses, estão algumas doenças cardiovasculares e, em particular, a insuficiência cardíaca. Cerca de 17 milhões de pessoas morrem de infarto do miocárdio ou insuficiência cardíaca por ano em todo o mundo ([Chicco & Jurman](https://bmcmedinformdecismak.biomedcentral.com/articles/10.1186/s12911-020-1023-5)).

Propomos, então, uma ferramenta para prever estatisticamente a probabilidade de sobrevivência a insuficiências cardíacas baseado em fatores metrificáveis de pacientes através de exames para guiar o processo de tomada de decisão médica quanto ao tratamento de um possível caso de insuficiência.

# Dados

Nos baseamos no trabalho de [Chicco & Jurman](https://bmcmedinformdecismak.biomedcentral.com/articles/10.1186/s12911-020-1023-5) para esse problema, que utiliza uma base de dados de 299 pacientes com insuficiência cardíaca coletados em 2015. Abaixo separamos os 13 atributos que cada paciente possui.

Atributos pessoais:
*  **age**: a idade em anos do paciente. Valores variam de 40 a 95 anos inteiros.
*  **sex**: se o paciente é do sexo feminino (representado por 0) ou masculino (representado por 1).

Atributos de doenças/hábitos:
* **smoking**: se o paciente é fumante (representado por 1) ou não (representado por 0).
*  **diabetes**: se o paciente possui diabetes (representado por 1) ou não (representado por 0).
*  **anaemia**: se o paciente possui anemia (representado por 1) ou não (representado por 0).
* **high_blood_pressure**: se o paciente possui hipertensão (representado por 1) ou não (representado por 0).

Atributos de exames de sangue/cardíacos:
*  **creatinine_phosphokinase**: concentração da enzima CPK  no sangue, medida com valores não negativos contínuos em microgramas/litro.
*  **ejection_fraction**: porcentagem de quanto do sangue deixa o coração a cada contração.
* **platelets**: quantidade de plaquetas no sangue, medida com valores não negativos contínuos em kiloplaquetas/mililitro.
* **serum_creatinine**: concentração de creatinina sérica no sangue, medida com valores não negativos contínuos em miligramas/decilitro.
* **serum_sodium**: concentração de sódio sérico no sangue, medida com valores não negativos contínuos em miliequivalentes/litro.

Atributos hospitalares:
* **time**: dias de acompanhamento passados. Valores variam de 4 a 285 dias inteiros.
* **DEATH_EVENT**: se o paciente sobreviveu após o período de acompanhamento (representado por 0) ou não (representado por 1).


# Tratamento de dados

Não há dados faltando em nenhum atributo. Entretanto, há um desbalanceamento com relação à classe objetivo, pois há mais dados de pacientes que sobreviveram do que de pacientes que pereceram. Realizamos, portanto, um balanceamento através de amostragens de tamanho equivalentes de ambos os conjuntos.

# Método de Solução por Naive Bayes

Uma solução para problemas de classificação, como ao qual estamos tratando neste, é utilizar Naive Bayes. Essa é uma técnica probabilística e pode ser definido a partir da ideia que dado um conjunto de variáveis aleatórias, variáveis de característica, onde os valores do conjunto é conhecido, o teorema quantifica a probabilidade condicional de uma variável aleatória. O termo Naive é devido ao classificador considerar que não há relação entre os objetos dentro da classe.

Sua eficiência está ligada a forma que o mesmo aprende. Já a velocidade traz custos, visto que fornece desempenho de generalização, o que é pior que alguns modelos lineares.

# Implementação do método

Para implementação será utilizada a biblioteca Sklearn. A mesma oferece 5 tipos de algoritmos para o Naive Bayes, sendo estes: Gaussian, Multinomial, Complement, Bernoulli e Categorical. As informações abaixo são derivadas da documentação da biblioteca.

* Multinomial: implementa o algoritmo Naive Bayes para dados multinomialmente distribuídos.
* Complement: é uma adaptação do algoritmo padrão multinomial naive Bayes (MNB) que é particularmente adequado para conjuntos de dados desbalanceados. Especificamente, o CNB usa estatísticas do complemento de cada classe para calcular os pesos do modelo.
* Bernoulli: adequado para dados que são distribuídos de acordo com distribuições multivariadas de Bernoulli; ou seja, pode haver vários recursos, mas cada um é assumido como uma variável de valor binário.
* Categorical: é adequado para classificação com características discretas que são categoricamente distribuídas. As categorias de cada recurso são extraídas de uma distribuição categórica.
* Gaussian: assume que a probalidade dos dados é gaussiana.


Portanto, para implementação, optamos pelo uso do algoritmo gaussiano por ser baseado em uma distribuição contínua, caracterizado por média e variância. Além de adequado para tarefas de classificação genéricas. [Link](https://www.springboard.com/blog/data-analytics/naive-bayes-classification/)

Foram feitos teste com dados normailizado e não normalizados. Resultados:
* Não normalizados: acurácia de 67%, recall de 58%, precision de 71% e F1 de 64%. 
* Normalizados: acurácia de 78%, recall 77%, precisão de 78% e F1 de 78%.

Com uso de validação cruzada, sendo 10, encontramos acurácia de 84% e 76% considerando a área embaixo da curva.

Um outro fator interessante é que foram realizdos teste para verificar se a escolha havia sido adequada, testando outros algoritmos. Para eles, a acurácia ficou abaixo de 60%, quando não balanceados. Já para balanceados, a acurácia ficou entre 65% e 60%, com exceção do Bernoulli que continuo abaixo.
