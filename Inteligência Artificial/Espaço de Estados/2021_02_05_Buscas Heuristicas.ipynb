# Resolução de problemas por meio de busca
## Buscas informadas (heurísticas)

- Consideram a probabilidade da escolha levar mais rapidamente à solução.
- Para alguns problemas, não há outras informações a considerar: usa-se uma busca cega.

## A melhor escolha

- Através de uma função de avaliação pode-se determinar qual nó deve ser expandido primeiro.
- Estabelece-se uma ordenação entre eles.
- Na verdade, é escolhido o nó que parece ser a melhor opção, de acordo com a avaliação.	

## Estratégias:

- **Busca gulosa**:  tenta minimizar o custo estimado para atingir a meta - o nó considerado mais próximo desta é escolhido primeiro. No  cálculo do custo estimado, usa-se uma  função heurística **h(n)** ( h(n)=0, se n é a meta ), a qual é específica para cada problema.
-  **Busca A* :** uma estratégia que combina a Busca Gulosa e a Busca de Custo Uniforme. Usa a função de avaliação **f(n) = g(n) + h(n)**, onde **g(n)**  dá o custo do início até um nó **n** e **h(n)** dá o custo estimado do menor caminho de **n** até a meta.
- **Busca A\* com Aprofundamento Iterativo (IDA\*)**: cada iteração é uma busca em profundidade. Usa-se uma **função de custo f** , ao invés de um limite de profundidade.
