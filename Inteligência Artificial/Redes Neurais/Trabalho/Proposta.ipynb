#Especificação do problema : Redes Neurais


A área médica é uma área especialmente desafiadora para o uso de técnicas de Inteligência Artificial. Essas técnicas podem ser aplicadas na solução dos mais diversos problemas dentro desse contexto. Um dos problemas de grande interesse é a previsão do óbito de pacientes internados. Desta forma, utilizando a base de dados [Patient Survival Prediction](https://www.kaggle.com/mitishaagarwal/patient?select=dataset.csv), projete e implemente uma rede neural que seja capaz de prever se um paciente que foi internado em um hospital irá morrer ou não. Deve ser feita uma análise criteriosa dos dados, a fim de determinar os hiperparâmetros da rede. Devem ser gerados gráficos que possam dar uma visão geral da base de dados. Os resultados da qualidade do modelo implementado devem estar explícitos nos *notebooks* a serem entregues para avaliação. Devem ser realizados diversos testes e a qualidade do modelo medida. 


## O trabalho

O trabalho deve ser feito em grupos, conforme definido. Cada grupo deverá entregar apenas um produto final para ser avaliado pelo professor. Apesar do trabalho ser feito em grupos, a avaliação de cada aluno será feita individualmente. Para este trabalho, os papéis devem ser trocados, ou seja, quem atuou com Engenheiro do Conhecimento no trabalho anterior deve assumir o papel de Analista de Dados, e vice-versa.


**Recomendações:**

   O primeiro passo para resolver um problema é entendê-lo muito bem, considerar o contexto e fazer as abstrações necessárias. A seguir, para não "reinventar a roda" é preciso identificar como a comunidade resolveu o problema em questão, ou problemas parecidos, cuja solução possa ser mapeada para situação em análise. Desta forma, deve ser feito um estudo, o mais amplo possível sobre esse assunto. O próximo passo é o grupo se reunir e instanciar o modelo em estudo para o caso específico do problema a ser resolvido. Em paralelo, uma análise rigorosa deve ser feita na base de dados a fim de prepará-la para seu uso no modelo. A seguir, com o modelo bem compreendido e instanciado e os dados devidamente preparados, a implementação do modelo pode ser feita. Finalmente, a solução produzida deve ser devidamente avaliada.
   
    

## Divisão de papéis:

* **Analista de dados:** deverá ficar responsável por fazer uma "Análise Exploratória de Dados" (AED), analisando a base de dados fornecida e preparando-a para ser usada na solução do problema. Esta análise deve produzir **diversos gráficos** que pemitam uma visão geral do comportamento dos dados. O produto final da análise é o conjunto de exemplos que servirá para o treinamento do modelo.

* **Engenheiro de conhecimento:** deverá produzir a solução para o problema, utilizando a técnica estudada, o modelo proposto pelo grupo e conjunto de exemplos preparado pelo analista de dados.



### Primeira etapa:

(1) **Notebook descrevendo detalhadamente o modelo:**
* Um modelo é uma representação abstrata do mundo. No nosso caso, o "mundo" é o contexto onde o problema se insere. Desta forma, deve ser analisado com cuidado o que deve e o que não deve ser considerado. Por exemplo, se a base de dados tem 30 colunas, todas elas deverão ser utilizadas na implementação? Algumas poderão ser desconsideradas? Por quê? Todas as decisões de projeto devem ser justificadas no notebook.
* Para utilizar a técnica em estudo, é preciso conhecê-la muito bem. Assim, nesta etapa, **deve ser feita uma descrição detalhada do problema e da técnica a ser utilizada.** A descrição do problema NÃO é a especificação apresentada pelo professor. O grupo deve pesquisar sobre o assunto e descrever o problema que será tratado, indicando os impactos. almejados pelo trabalho
* Deve ser elaborado um modelo com base na técnica em estudo.

### Segunda etapa:

(2) **Notebook contendo a Análise Exploratória dos Dados**: deve ser entregue pelo Analista de Dados.

(3) **Notebook contendo a solução para o problema**: deve ser entregue pelo Engenheiro de Conhecimento.

(4) **Uma apresentação explicando detalhadamente como a técnica estudada foi utilizada para resolver o problema proposto**. Esta NÃO deve ser uma apresentação dos *notebooks* nem deve apresentar detalhes de codificação.. A apresentação deve trazer uma descrição do problema, do modelo utilizado e da solução desenvolvida, de preferência utilizando slides ou algo parecido.
