# Previsão de Sobrevivência do Paciente

A probabilidade de sobrevivência de um paciente é um fator extremamente importante, inclusive para os familiares, por fornecê-lhes esperança. Já para os profissionais a análise de sobrevivência é como um fator de tempo. Quanto tempo até a
ocorrência de um evento ou o risco de ocorrência de um
evento por unidade de tempo. 

Diversos fatores podem influenciar a estabilidade de um paciente. Para o pior dos casos, levando este a óbito.

# Dados

Os dados recebidos são originados da plataforma Kagle, contendo 85 atributos e 91713 informações sobre internações de pacientes, deste 7915 são óbitos e 83797 não óbitos. Resultando que 8% das internações, levam a morte.

Os atributos podem ser dividos em:
*  identificação, seja do hospital, do paciente ou da unidade de admissão deste; 
*  presença de doenças, como AIDS, Leucemia, Cirrose, etc; 
*  informações sobre pressão, nível de glicose e potássio;
*  informações descritivas sobre o paciente, como, peso, altura, idade, massa corporal, sexo, etnia, se o mesmo deu entrada para cirurgia, peso;
*  parâmetros para escala APACHE e os scores desta;


# Método de Solução por Redes Neurais

Uma solução para problemas de classificação, como o da previsão de sobrevivência do paciente, é utilizar redes neurais artificiais. Essa técnica se baseia no funcionamento do cérebro humano, dividido em pequenas unidades chamadas neurônios, que são estimulados por seus vizinhos e, dependendo do nível do estímulo, produzem estímulos em outros vizinhos. 

A estrutura de um neurônio artificial se divide em:

*   Entradas: valores que são recebidos pelo neurônio.
*   Pesos: valor pelo qual multiplicamos uma entrada.
*   Função de ativação: combina (soma) valores de entrada já pesados e aplica uma função para definir a saída. Os tipos mais comuns de funções de ativação são:
  *   Linear: o valor resultante da combinação é passado diretamente como saída.
  *   Limiar: o valor resultante da combinação é comparado com um limite. Se for menor que ele, a saída é 0, senão, é 1 (ou outro número qualquer maior que 0).
  *   Sigmoidal: uma função diferenciável e estritamente crescente (com formato similar a um S deitado) aplicada ao valor resultante da combinação.



A rede neural artificial se estrutura ligando saídas de neurônios como entradas de outros. Dividimos eles em camadas, como na figura abaixo (Katti Faceli). Cada atributo de entrada é conectado a todos os neurônios da primeira camada intermediária (ou diretamente à camada de saída, caso não haja camada intermediária). A camada de saída corresponde ao valores a serem previstos. Para problemas de regressão (previsão de um número real), a camada de saída possui um neurônio para cada atributo a ser previsto. Para problemas de classificação, cada neurônio da camada de saída representa uma classe. A classe que possuir maior saída é aquela à qual a rede diz que o objeto pertence.

A topologia de uma rede neural artificial é definida pelos seguintes atributos:

*   Função de ativação escolhida: usada em todos os neurônios da rede.
*   Número de camadas intermediárias: camadas antes da camada de saída.
*   Número de neurônios: pode ser diferente para cada camada.
*   Grau de conectividade: como os neurônios de uma camada se conectam à camada seguinte. Normalmente utilizamos a abordagem "completamente conectada", que liga todos os neurônios da camada j à camada j+1. Optamos por essa abordagem por ser difícil dizer qual ligação pode ser retirada sem prejudicar a previsão.
*   Backpropagation: numa rede com backpropagation, podemos ligar a saída de um neurônio como uma entrada dele mesmo, ou como a entrada de um neurônio de uma camada anterior. Essa solução é especialmente indicada para problemas em que há uma sequenciação entre atributos ou para soluções em sistemas dinâmicos.




Após definir a topologia da rede neural, temos já seu modelo "estático". Falta apenas colocá-la para treinar e aprender com o conjunto de dados de treino. Como quase todos os atributos da rede neural já foram definidos, o aprendizado é definido apenas pelo valor de cada peso que multiplica cada entrada de cada neurônio. Para o problema da previsão da sobrevivência de pacientes, como já temos a resposta correta para cada entrada, o método de aprendizado é supervisionado e funciona através da solução de um problema de minimização de erro a cada entrada recebida pela rede.


Por fim, há ainda a adaptabilidade de um algoritmo de aprendizagem de minimização por erros. Podemos modificar o quanto mudamos os pesos a cada objeto recebido através da multiplicação dessa alteração por uma constante Alfa. Um Alfa muito grande pode acelerar o processo de aprendizagem, porém dificultar a convergência do modelo para uma solução. Já um Alfa baixo, garante uma convergência mais precisa, porém exige muito mais ciclos de treinamento.

# Implementação do método

Ao utilizar como ferramenta o Google Collab, há algumas limitações de memória e processamento que nos impedem de expandir muito os testes em modelos. Para lidar com essa situação, definimos as seguintes limitações:

*   Máximo de 3 camadas.
*   Máximo de 30 neurônios por camadas.

Para evitar perda de informações de um neurônio a outro, decidimos pelo grau de conectividade completo. A rede conta com backpropagation para aplicar a descida de gradiente, bem como a função de ativação sigmoidal pelo mesmo motivo.

O método de aprendizado escolhido na ferramenta é conhecido como 'adam', que utiliza o gradiente da função de ativação para minimizar o erro. É um método mais teoricamente estudado e conhecido em livros de inteligência artificial, então nos interessa pela sua consolidação. Por fim, utilizamos também uma constante Alfa relativamente baixa para maior convergência dos resultados.

Nossos testes consistem, portanto, de exaustivamente treinar combinações distintas de camadas e neurônios para chegar numa solução mais precisa para o problema.
