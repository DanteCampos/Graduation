## Transferindo os dados para planilha

```python
import pandas as pd
 
# Cria dataframes do pandas a partir dos csvs
fat = pd.read_csv('Fat_Supply_Quantity_Data.csv')
kcal = pd.read_csv('Food_Supply_kcal_Data.csv')
kg = pd.read_csv('Food_Supply_Quantity_kg_Data.csv')
protein = pd.read_csv('Protein_Supply_Quantity_Data.csv')
```

```python
# Trocando legendas das colunas da tabela de gorduras
columns = list(fat.columns)

mapFat = {}
mapKg = {}
mapKcal = {}
mapProtein = {}

for i in range(1, 25):
  mapFat.update({columns[i] : "Fat - " + columns[i]})
  mapKg.update({columns[i] : "Kg - " + columns[i]})
  mapKcal.update({columns[i] : "Kcal - " + columns[i]})
  mapProtein.update({columns[i] : "Protein - " + columns[i]})

fat.rename(mapper = mapFat, axis = 1, inplace = True)
kg.rename(mapper = mapKg, axis = 1, inplace = True)
kcal.rename(mapper = mapKcal, axis = 1, inplace = True)
protein.rename(mapper = mapProtein, axis = 1, inplace = True)
```

```python
# Dropando dados redundantes ou desnecessários
dropLabels = columns[25:32]

fat.drop(labels = columns[31], axis = 1, inplace = True)
kg.drop(labels = dropLabels, axis = 1, inplace = True)
kcal.drop(labels = dropLabels, axis = 1, inplace = True)
protein.drop(labels = dropLabels, axis = 1, inplace = True)
```

```python
# Unindo as tabelas de acordo com o país

df = pd.merge(kg, kcal)
df = pd.merge(df, protein)
df = pd.merge(df, fat)
```

# Implementação Algoritmo Genético

Os pesos considerados serão mortes menor que a média (0.039370) ou recuperação maior que a média(1.452356).

```python
!pip install pyeasyga
```

```python
from pyeasyga import pyeasyga
```

```python
ga = pyeasyga.GeneticAlgorithm(data)        # initialise the GA with data
```

```python
# define a fitness function
def fitness(individual, data):
    media_proteina = []
    proteina = 0

    for selected, dieta in zip(individual, data):
        if selected and (dieta['Deaths'] < 0.039370 or dieta['Recovered'] > 1.452356):
          proteina += dieta['Protein - Alcoholic Beverages'] + dieta['Protein - Animal Products'] + dieta['Protein - Animal fats']  + dieta['Protein - Aquatic Products, Other']
          proteina += dieta['Protein - Cereals - Excluding Beer'] + dieta['Protein - Eggs'] + dieta['Protein - Fish, Seafood']  + dieta['Protein - Fruits - Excluding Wine'] + dieta['Protein - Meat']
          proteina += dieta['Protein - Milk - Excluding Butter'] + dieta['Protein - Offals'] + dieta['Protein - Oilcrops']  + dieta['Protein - Pulses'] + dieta['Protein - Spices']
          proteina += dieta['Protein - Starchy Roots'] + dieta['Protein - Sugar Crops'] + dieta['Protein - Sugar & Sweeteners']
          proteina += dieta['Protein - Stimulants'] + dieta['Protein - Treenuts'] + dieta['Protein - Vegetal Products']  + dieta['Protein - Obesity']
          proteina += dieta['Protein - Vegetable Oils'] + dieta['Protein - Vegetables'] + dieta['Protein - Miscellaneous']

        media_proteina.append(proteina/24)

    return sum(media_proteina)/len(media_proteina)

ga.fitness_function = fitness               # set the GA's fitness function
```

```python
ga.run()                                    # run the GA
```

```python
#Melhor individuo considerando proteina
print(ga.best_individual())                  # print the GA's best solution
```
